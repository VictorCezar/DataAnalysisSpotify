```python
#Músicas mais longas
songs_by_duration = dados.sort_values("min:sec", ascending=False)[["song", "artist", "min:sec"]]
songs_by_duration[:5]
```

```python
#Músicas mais curtas
songs_by_duration[-5:]
```

```python
songs_by_duration_short = dados.sort_values("min:sec", ascending=True)[["song", "artist", "min:sec"]]
songs_by_duration_short[:5]
```

```python
# Top 5 músicas por popularidade
Popular_songs = dados.sort_values(['popularity'],axis=0,ascending=False)
Popular_songs[:5]
```

# **Identificando e Removendo Outliers**

```python
plt.rc('figure', figsize = (10, 8))
dados.boxplot(['danceability'])
```

```python
dance = dados['danceability']
```

```python
Q1 = dance.quantile(.25)
Q3 = dance.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (dance >= limite_inferior) & (dance <= limite_superior)
dados_new = dados[selecao]
```

```python
dados_new.boxplot(['danceability'])
```

```python
dados.hist(['danceability'])
dados_new.hist(['danceability'])
```
